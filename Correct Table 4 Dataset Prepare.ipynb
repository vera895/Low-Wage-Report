# dummy variable

```python
data = q1.loc[:, ['SEX','AGE','MARST','RACE','EDUC','IND','EMPSTAT','OCC','OCCLY','INDLY']]
```

```python
data
```

```python
data['male'] = data['SEX']
data.loc[data['male'] != 1 , 'male'] = 0

data['age'] = data['AGE']

data['Married'] = data['MARST']
data.loc[data['Married'].isin([1,2]),'Married'] =1
data.loc[data['Married']!= 1, 'Married'] = 0

#white
data['White'] = data['RACE']
data.loc[data['RACE'] == 100, 'White'] =1
data.loc[data['White'] !=1, 'White'] = 0

# Non-White
data['Non_white'] = data['RACE']
data.loc[data['Non_white']!= 100, 'Non_white'] = 1
data.loc[data['Non_white'] !=1, 'Non_white'] = 0
```

```python
data['education'] = data['EDUC']
#four_year_college
data.loc[data['EDUC'].isin([123,111,125,124]),'education']=4
# som_college
data.loc[data['EDUC'].isin([80,90,100,110,120,122,81,92,91]),'education']=3
#high_school
data.loc[data['EDUC']==73,'education']=2
# less than high school
data.loc[~data['education'].isin([4,2,3]),'education']=1
```

# merge Industry

```python
data['ind1'] = data['IND']
# industry replace 0 with INDLY
data.ind1[data.ind1 == 0] = data.INDLY
```

```python
data.ind1.value_counts().sort_index()
```

```python

data.loc[(data['ind1'] <= 291) & (data['ind1'] >1), 'ind1'] = 1
data.loc[(data['ind1'] >=370) & (data['ind1'] < 491), 'ind1'] = 2
data.loc[(data['ind1'] >=570) & (data['ind1'] < 691), 'ind1'] = 3
data.loc[data['ind1'] == 770, 'ind1'] = 4
data.loc[(data['ind1'] >=1070) & (data['ind1'] < 3991), 'ind1'] = 5
data.loc[(data['ind1'] >=4070) & (data['ind1'] < 4591), 'ind1'] = 6
data.loc[(data['ind1'] >=4670) & (data['ind1'] < 5791), 'ind1'] = 7
data.loc[(data['ind1'] >=6070) & (data['ind1'] < 6391), 'ind1'] = 8
data.loc[(data['ind1'] >=6470) & (data['ind1'] < 6791), 'ind1'] = 9
data.loc[(data['ind1'] >=6870) & (data['ind1'] < 7191), 'ind1'] = 10
data.loc[(data['ind1'] >=7270) & (data['ind1'] < 7791), 'ind1'] = 11
data.loc[(data['ind1'] >=7860) & (data['ind1'] < 8471), 'ind1'] = 12
data.loc[(data['ind1'] >=8560) & (data['ind1'] < 8691), 'ind1'] = 13
data.loc[(data['ind1'] >=8770) & (data['ind1'] < 9291), 'ind1'] = 14
data.loc[(data['ind1'] >=9370) & (data['ind1'] < 9591), 'ind1'] = 15
data.loc[(data['ind1'] >=9670) & (data['ind1'] < 9891), 'ind1'] = 16
```

# move result
