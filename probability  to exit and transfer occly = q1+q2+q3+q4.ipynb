```python
#four_year_college
df.loc[df['EDUC'].isin([123,111,125,124]),'education']=4
# som_college
df.loc[df['EDUC'].isin([80,90,100,110,120,122,81,92,91]),'education']=3
#high_school
df.loc[df['EDUC']==73,'education']=2
# less than high school
df.loc[~df['education'].isin([4,2,3]),'education']=1
```

```python
# race
df.loc[df['RACE'] == 100, 'race2'] =1
df.loc[df['RACE'] == 200, 'race2'] = 2
df.loc[df['RACE'] == 651, 'race2'] = 3
df.loc[~df['RACE'].isin([100,200,651]), 'race2'] = 4
```
