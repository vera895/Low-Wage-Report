# Analyze Q1-Q4 Demographic

```python
q1_ob['quartile']='q1'
q2_ob['quartile']='q2'
q3_ob['quartile']='q3'
q4_ob['quartile']='q4'
```

```python
data = pd.concat([q1_ob, q2_ob, q3_ob, q4_ob])
```

```python
data
# 74707 rows Ã— 54 columns
```

```python
edu = data.groupby(['quartile'])['EDUC'].value_counts().sort_index()
edu = pd.DataFrame(edu)
edu.columns = ['num_of_employee']
edu = edu.reset_index()
edu.loc[edu['EDUC']==73, 'education']='High School'
edu.loc[edu['EDUC'].isin([80,90,100,110,120,122,81,92,91]), 'education']='Some College'
edu.loc[edu['EDUC'].isin([123,111,125,124]), 'education']='Four-Year College Degree or More'
edu.loc[~edu['EDUC'].isin([73,80,90,100,110,120,122,81,92,91,123,111,125,124]), 'education']='Less Than High School'
```

```python
edu
```

```python
edu2 = edu.groupby(['quartile','education'])['num_of_employee'].sum()
```

```python
edu2 = pd.DataFrame(edu2)
edu2
```

```python
edu2.num_of_employee.sum()
```

```python
# marital status
mar = data.groupby(['quartile'])['MARST'].value_counts().sort_index()
mar = pd.DataFrame(mar)
mar.columns = ['num_of_employee']
mar = mar.reset_index()
```

```python
mar
```

```python
mar.loc[mar['MARST'].isin([1,2]), 'marriage']='Married'
mar.loc[mar['MARST'].isin([3,4,5]), 'marriage']= 'Separated/Divorced/Widowed'
mar.loc[mar['MARST']==6, 'marriage']='Never married/single'
```

```python
mar
```

```python
mar2 = mar.groupby(['quartile', 'marriage'])['num_of_employee'].sum()
mar2 = pd.DataFrame(mar2)
mar2
```

```python
mar2.num_of_employee.sum()
```

```python
# race
race = data.groupby(['quartile'])['RACE'].value_counts().sort_index()
race = pd.DataFrame(race)
race.columns = ['num_of_employee']
race = race.reset_index()
race.loc[race['RACE']== 100, 'race2']='White'
race.loc[race['RACE']== 200, 'race2']= 'Black'
race.loc[race['RACE']== 651, 'race2']= 'Asian'
race.loc[~race['RACE'].isin([100,200,651]), 'race2'] ='Other Races'
race2 = race.groupby(['quartile', 'race2'])['num_of_employee'].sum()
race2 = pd.DataFrame(race2)
race2
```

```python
race2.num_of_employee.sum()
```

```python
# Sex
gender = data.groupby(['quartile'])['SEX'].value_counts().sort_index()
gender = pd.DataFrame(gender)
gender.columns = ['num_of_employee']
gender = gender.reset_index()
gender.loc[gender['SEX']== 1, 'gender2']='Male'
gender.loc[gender['SEX']== 2, 'gender2']= 'Female'

gender2 = gender.groupby(['quartile', 'gender2'])['num_of_employee'].sum()
gender2 = pd.DataFrame(gender2)
gender2
```

```python
gender2.num_of_employee.sum()
```

```python
# AGE
age = data.groupby(['quartile'])['AGE'].value_counts().sort_index()
age = pd.DataFrame(age)
age.columns = ['num_of_employee']
age = age.reset_index()
```

```python
age
```

```python
bins = np.array([17,20,30,40,50,60])
groups = age.groupby(('quartile',pd.cut(age.AGE, bins)))
out = groups.sum().dropna()
```
