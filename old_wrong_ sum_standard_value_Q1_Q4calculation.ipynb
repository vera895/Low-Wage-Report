```python
import pandas as pd
obs_ten = pd.read_csv('C:/Users/yfeng/Desktop/All_obs_10_presitige.csv')
```

```python
df = obs_ten.loc[~obs_ten['EARNWEEK'].isin([9999.99,0])]
df = df.loc[~df['UHRSWORKT'].isin([997,999,0])]

df['Average_Weekly_Hours_Compareed_to_40'] = abs(df['UHRSWORKT'] - 40)

df1 = df.loc[df['UHRSWORKT'] > 40]
df2 = df.loc[df['UHRSWORKT'] <= 40]

df1['Calcu_Hourly_Wage'] = df1['EARNWEEK']/(1.5*df1['UHRSWORKT']-20)
df2['Calcu_Hourly_Wage'] = df2['EARNWEEK']/df2['UHRSWORKT']

df = pd.concat([df1,df2])


df = df[df['PAIDGH'] != 0.0]
df = df[pd.notna(df['PAIDGH'])]
df['Employer_paid_inur'] = df['PAIDGH']
```

```python
df
```

```python
df['Employer_paid_inur2'] = df['Employer_paid_inur']
df.loc[df['Employer_paid_inur2'].isin([21,22]), 'Employer_paid_inur2']=1
df.loc[df['Employer_paid_inur2'].isin([10]), 'Employer_paid_inur2']=0
df
```

```python
dc1 = df[pd.notna(df['PRESGL'])]
dc1
```

```python
dc2 = dc1.iloc[:,-6:]
cols = list(dc2.columns)
cols.remove('PRENT')
cols.remove('Employer_paid_inur')
aa = dc2[cols]
aa
```

```python
from scipy.stats import zscore
aa2 = aa.apply(zscore)
aa2
```

```python
aa3 = pd.concat([aa,aa2], axis=1)
aa3.columns = ['PRESGL','Average_Weekly_Hours_Compareed_to_40','Calcu_Hourly_Wage','Employer_paid_inur2', 
              'PRESGL_zscore','Hours_to_40_zscore','Hourly_Wage_zscore','Employer_paid_inur2_zscore']
aa3
```

```python
index1 = pd.concat([dc1,aa3], axis=1)
index1
```

```python
index1['index_score'] = index1.PRESGL_zscore + index1.Hours_to_40_zscore + index1.Hourly_Wage_zscore + index1.Employer_paid_inur2_zscore
index1
```

```python
index1.to_csv('C:/Users/yfeng/Desktop/cps/Sep_17_new_Q1/Index_score.csv')
```

```python
occ_index_score = index1.groupby(['OCC'])['index_score'].mean()
occ_index_score = pd.DataFrame(occ_index_score)
occ_index_score
```

```python
occ_index_score.index_score.quantile([0.25,0.5,0.75,1.0])
```
